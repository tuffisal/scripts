Função relplot: método para visualização de relações estatísticas entre variáveis. Combina o scatter plot e line plots
Muito útil para visualizar relações entre variáveis

```python
#Visualizando a relação entre o valor total da conta e a gorjeta
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
tips.count()
```

```python
#Visualizando a relação entre as variáveis total da conta e gorjeta e pintando de cores diferentes os fumantes
sns.relplot(x='total_bill',y='tip',hue='smoker',data=tips) #hue = matiz
```

```python
#Visualizando a relação entre as variáveis total da conta e gorjeta e pintando de cores diferentes a variável sexo 
sns.relplot(x='total_bill',y='tip',hue='sex',data=tips)
```

```python
#Visualizando a relação entre as variáveis total da conta e gorjeta e pintando de cores diferentes a variável sexo e definindo um estilo para variável sexo
sns.relplot(x='total_bill',y='tip',hue='sex',style='sex',data=tips)
```

Representando quatro variáveis - tomar cuidado para não deixar a visualização confusa

```python
sns.relplot(x='total_bill',y='tip',hue='sex',style='smoker',data=tips)
```

Em caso de variáveis numéricas para definir a tonalidade a paleta de cores altera de forma sequencial

```python
sns.relplot(x='total_bill',y='tip',hue='size',data=tips)
```

```python
sns.set_palette(sns.color_palette('Blues_d'))
sns.relplot(x='total_bill',y='tip',hue='size',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',size='size',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',size='size',sizes=(15,300),data=tips)
```

Visualização a relação entre as 4 variáveis total_bill,tipo,day,smoker

```python
%matplotlib notebook
sns.set()
sns.scatterplot(x='total_bill',y='tip',hue='day',style='time',data=tips)
```

Redimensionando figuras

```python
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(15,7))
sns.scatterplot(x='total_bill',y='tip',hue='day',data=tips)
```

```python
#Paleta de cores para dados quantitativos
sns.palplot(sns.color_palette('Accent'))
```

```python
#Paleta de cores para dados divergentes
sns.palplot(sns.color_palette('PuOr'))
```

```python
#Paleta de cores para dados sequenciais
sns.palplot(sns.color_palette('Blues'))
```

```python
#Gráfico usando a paleta de cor sequencial Blues
sns.relplot(x='total_bill',y='tip',hue='size',palette='Reds',size='size',sizes=(12,250),data=tips)
```

Paleta de cores qualitativa

```python
sns.catplot(x='sex',kind='count',palette='Set2',data=tips)
```

Paleta de cores divergente

```python
sns.heatmap(tips.corr(),annot=True,cmap='PiYG')
```

Definindo cores padrão para gráficos Seaborn

Instrução with para definir cor temporariamente

```python
with sns.color_palette('Accent'):
    sns.relplot(x='total_bill',y='tip',size='size',sizes=(12,250),data=tips)
```
