```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
iris = pd.read_csv("datasets\iris.csv")
```

```python
iris.head()
```

```python
X = iris.drop('Species',axis=1)
```

```python
iris.head()
```

```python
X.head()
```

```python
Y = iris.Species
```

```python
Y.head()
```

```python
Y.unique()
```

```python
iris.Species.unique()
```

Convertendo valores categóricos de classe em numéricos

```python
def converte_classe(l):
    if l == 'Iris-virginica':
        return 0
    elif l == 'Iris-setosa':
        return 1
    elif l == 'Iris-versicolor':
        return 2
```

```python
Y.unique()
```

```python
Y = Y.apply(converte_classe)
```

```python
Y.value_counts()
```

```python
Y.head()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'random')
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
distance = kmeans.fit_transform(X)
```

```python
distance
```

```python
distance[0]
```

```python
%matplotlib notebook
x = ['Cluster 0', 'Cluster 1', 'Cluster 2']
plt.barh(x,distance[0])
plt.xlabel('Distância')
plt.title('Distância por Clusters')
plt.show()
```

```python
labels = kmeans.labels_
```

```python
labels
```

```python
%matplotlib notebook
plt.figure(figsize=(8,6))
plt.scatter(X['SepalLength'], X['SepalWidth'], s=100, c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red', label='Centroids')
plt.title('Dataset Iris e Centroids')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.show()
```
